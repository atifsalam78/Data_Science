```python
def reps_sales(sales):
    total_sales = sales["Sales"].sum()
    total_per = 0
    
    for rep in sales["Name of Sales Rep"].unique():
        total_rep_sales = sales[sales["Name of Sales Rep"] == rep]["Sales"].sum()
        per_sales = (total_rep_sales / total_sales)*100
        total_per += per_sales
        formated_rep_sales = f"{total_rep_sales: ,.2f}"    
        print(f"{rep}: {formated_rep_sales}, Percentange: {per_sales: ,.2f}%")
    print(f"{total_per: ,.2f}%")        
```

```python
reps_sales(sales)
```

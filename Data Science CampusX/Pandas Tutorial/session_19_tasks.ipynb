### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
# code here

import pandas as pd
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQjh5HzZ1N0SU7ME9ZQRzeVTaXaGsV97rU8R7eAcg53k27GTstJp9cRUOfr55go1GRRvTz1NwvyOnuh/pub?gid=1562145139&single=true&output=csv"
titanic_df = pd.read_csv(url)

p_class = titanic_df.groupby("Pclass")
p_class["Embarked"].transform(lambda x: x.fillna(x.mode().iloc[0]))

group_embarked_counts = p_class["Embarked"].value_counts().sort_index()
print(group_embarked_counts)
````

````python
# Create a function that can return the highest score of any batsman
def highest_score(batsman):
    df = ipl[ipl["batsman"] == batsman]
    match_number = df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).index[0]
    score = df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
    print("'{}' scored his highest '{}' in match number '{}'".format(batsman, score, match_number))
````

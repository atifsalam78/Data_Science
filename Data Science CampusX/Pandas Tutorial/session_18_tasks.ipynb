###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
def point_table(season):
    match_dict = {"MatchesPlayed": match_dataset.loc[match_dataset["Season"]== season, "Team1"].value_counts()+
              match_dataset.loc[match_dataset["Season"]== season, "Team2"].value_counts(),
             "MatchesWon": match_dataset.loc[match_dataset["Season"] == season, "WinningTeam"].value_counts(),
              "NoResults": match_dataset[match_dataset["WonBy"] == "NoResults"]["Team1"].value_counts().add(match_dataset[match_dataset["WonBy"] == "NoResults"]["Team2"].value_counts(), fill_value=0),
             "Points": match_dataset.loc[match_dataset["Season"] == season, "WinningTeam"].value_counts()*2}
    point_df = pd.DataFrame(match_dict)
    point_df.fillna(0, inplace=True)
    point_df["Points"] = point_df["Points"].add(point_df["NoResults"])
    return point_df.sort_index()

point_table("2019")          
````

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
relevant_df = match_dataset[["MatchNumber", "Team1", "Team2", "WinningTeam"]]
lost_teams = []
for index, row in relevant_df.iterrows():
    match_number = row["MatchNumber"]
    team1 = row["Team1"]
    team2 = row["Team2"]
    winning_team = row["WinningTeam"]
    
    lost_team = team1 if winning_team != team1 else team2
#     season_position = "Winner" if winning_team in match_dataset["Team1"] else "Runner"
    
    season_position = "Winnner" if winning_team == match_dataset.at[index, "Team1"] else "Runner"
    lost_teams.append({"MatchNumber": match_number, "Team1":team1, "Team2": team2,
                       "WinningTeam":winning_team, "LostTeam": lost_team,
                      "SeasonPosition": season_position})
    
lost_teams_df = pd.DataFrame(lost_teams)

lost_teams_df[["MatchNumber","Team1", "Team2", "WinningTeam", "LostTeam", "SeasonPosition"]]
````

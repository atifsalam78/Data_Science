````python
import matplotlib.pyplot as plt

# Assuming 'price' is a DataFrame with 'course_name' as index and 'price' as the column
# You can reset the index to have 'course_name' as a regular column for plotting

price = courses.merge(regs, how="inner", on="course_id").groupby("course_name")[["price"]].sum().sort_values(by="price")
price.reset_index(inplace=True)

# Plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(price['course_name'], price['price'], color='skyblue', edgecolor='black', linewidth=1.2)

# Adding data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom', fontsize=10)

# Adding labels and title
plt.xlabel('Course Name', fontsize=14)
plt.ylabel('Total Price', fontsize=14)
plt.title('Total Price of Each Course', fontsize=16)

# Adding grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Adding a legend
plt.legend(['Total Price'], loc='upper right')

# Adding a background color
plt.axhspan(0, price['price'].max(), facecolor='lightgray', alpha=0.5)

# Adding a horizontal line at the average price
avg_price = price['price'].mean()
plt.axhline(avg_price, color='red', linestyle='--', linewidth=2, label=f'Average Price: {round(avg_price, 2)}')

# Adding a text annotation for the average price
plt.text(0.02, 0.95, f'Average Price: {round(avg_price, 2)}', transform=plt.gca().transAxes, color='red',
         fontsize=12, verticalalignment='top')

# Adjusting layout
plt.tight_layout()

# Show the plot
plt.show()
````

# Class Revision

### Concat

* In `Vertical` stacking number of columns must be same.
```python
    pd.concat([nov, dec], ignore_index=True)
```
* In `Horizontal` stacking no requirment of same number of columns.
```python
    pd.concat([nov, dec], axis=1)
```

* Problem in concat is that it will retain index.

### Merge

Join two or more data frame on the basis of one common column.

#### Inner Join

It will display only common column's data, this is default.

```python
df1.merge(df2, how="inner", on="student_id")
```

#### Full Join

It will display all data
```python
df1.merge(df2, merge="outer", on="student_id")
```

#### Left Join

It will display left data frame's data irrespective of data is available in right data frame.

```python

df1.merge(df2, merge="left", on="student_id")
```

#### Right Join

It will display right data frame's data irrespective of data is available in left data frame.

```python
df1.merge(df2, merge="right", on="student_id")
```

#### Alternate Syntax
```python
pd.merge(df1, df2, how="inner", on="student_id")
```

### Multi Index Series & Data Frame

* Series is 1D and Data Frames are 2D object.
* To fetch data from series we required only 1 piece of information i.e. (index).
* To fetch data from data frame we required 2 piece of information i.e. (index, column).
* Multi index series also know as hierarchical indexing.
* ```unstack()``` convert multi index series into data frame i.e. row to column.
* ```stack``` convert data frame into series i.e. column to row, it will apply only on data frame not on series.

#### What is the need of Multi Indexing?

To represent higher dimension (2D, 3D or 10D) data into lower dimension, and this is the true power of ```pandas```.

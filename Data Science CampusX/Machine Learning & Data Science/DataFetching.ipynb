```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
final = pd.DataFrame()
for j in range(1,11):
    url = "https://www.businesslist.pk/location/karachi/{}".format(j)
    webPage=requests.get(url).text
    soup = BeautifulSoup(webPage,"lxml")
    company = soup.find_all("div", class_="company with_img g_0")
    name = []
    address = []
    verify = []
    rating = []
    # old = []

    for i in company:
        name.append(i.find("h4").text.strip())
        address.append(i.find("div",class_="address"))
        verify.append(i.find("u", class_="v"))
        rating.append(i.find("div", class_="rate"))
    
    d = {"name":name,"address":address,"verify":verify,"rating":rating}
    df = pd.DataFrame(d)
final = final.append(df, ignore_index=True)
```

```python
import matplotlib.pyplot as plt
x = [5, 7, 8, 7, 2, 11, 12, 9, 6, 11, 12, 9, 6]
y = [99, 86, 87, 111, 86, 103, 87, 94, 78, 77, 85, 96, 99]

plt.scatter(x, y)
plt.show()
```

```python
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
def func(x):
    return slope*x + intercept

mymodel = list(map(func, x))
plt.scatter(x, y)
plt.plot(x, mymodel)

plt.show()
```

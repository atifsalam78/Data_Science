```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Location:
    def __init__(self, location_x, location_y, destination_x, destination_y):
        self.location = Point(location_x, location_y)
        self.destination = Point(destination_x, destination_y)

    def reflect_on_x_axis(self):
        reflected_y = -self.destination.y
        reflected_point = Point(self.destination.x, reflected_y)
        print(f"Reflection of destination: ({reflected_point.x}, {reflected_point.y})")


# Example usage
location = Location(2, 3, 4, 5)
location.reflect_on_x_axis()
```

```python
class Atm:
    
    def __init__ (self):
        self.pin = ""
        self.balance = 0
        self.menu()
        
    
    def menu(self):
        user_input = input(""" 
        1. Press 1 for create pin
        2. Press 2 for change pin
        3. Press 3 for check balance
        4. Press 4 for withdraw
        5. Anything else to exit
        """)
    
        if user_input == "1":        
            self.create_pin()
    
        elif user_input == "2":
            self.change_pin()        
    
        elif user_input == "3":
              self.check_balance()        
    
        elif user_input == "4":
              self.withdraw()
                
        else:
            exit()
        
    def create_pin(self):
        user_pin = input("Create your pin: ")
        user_balance = int(input("Enter your balance: "))
        
        self.pin = user_pin
        self.balance = user_balance
        print("Your pin created successfully!")
        
        self.menu()
        
    def change_pin(self):
        user_pin = input("Enter existing pin: ")
        
        if user_pin == self.pin:
            new_pin = input("Enter your new pin: ")
            self.pin = new_pin
            print("Your pin changed successfully!")
        
        else:
            print("provided pin is not authorized to do so!")
        
        self.menu()    
    
    def check_balance(self):
        user_pin = input("Enter your pin: ")
        
        if user_pin == self.pin:
            print("Your current balance is {}".format(self.balance))
        else:
            print("provided pin is not authorized to do so!")
            
        self.menu()
    
    def withdraw(self):
        user_pin = input("Enter your pin: ")
        
        if user_pin == self.pin:
            user_amount = int(input("Enter amount to be withdraw: "))
            
            if user_amount <= self.balance:
                self.balance = self.balance - user_amount
                print("{} withdrawn successfully, your current balance is {}".format(user_amount, self.balance))
            else:
                print("Insufficient amount")
                
        else:
            print("provided pin is not authorized to do so!")
            
        self.menu()
```

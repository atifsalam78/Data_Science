```python
def cumalitive_cases(country):
    
    bol = who_df["Country"] == "{}".format(country)
    country_name = who_df[bol]
    plt.plot(country_name["Date_reported"], country_name["Cumulative_cases"])
    plt.title("{} - Covid Commulative Cases".format(country))
    plt.xlabel("Date Reported")
    plt.ylabel("Comulative Cases")
    plt.show
    
cumalitive_cases("Italy")
```

```python
def new_cases(country):
    
    bol = who_df["Country"] == "{}".format(country)
    country_name = who_df[bol]
    plt.plot(country_name["Date_reported"], country_name["New_cases"])
    plt.title("{} - Covid New Cases".format(country))
    plt.xlabel("Date Reported")
    plt.ylabel("New Cases")
    plt.show
    
new_cases("Italy")
```

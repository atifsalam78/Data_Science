```python
# To unzip zip files

import zipfile
with zipfile.ZipFile("bike.zip") as zipped:
    zipped.extractall("bikeshare")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(bike.dteday, bike.cnt)
plt.show()
# This data available for both casual and regular riders
```

```python
plt.plot(bike.dteday, bike.casual, label = "Casual")
plt.plot(bike.dteday, bike.registered, label = "Regsitered")
plt.show()
```
